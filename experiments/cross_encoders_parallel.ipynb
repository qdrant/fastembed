```python
from fastembed.rerank.cross_encoder import TextCrossEncoder
import numpy as np
model_name = """Xenova/ms-marco-MiniLM-L-12-v2"""
model = TextCrossEncoder(model_name=model_name)

query = "What is the capital of France?"
documents = ["Paris is the capital of France.", "Berlin is the capital of Germany."]
scores = np.array(list(model.rerank(query, documents)))
scores
```

```python
pairs = [("What is the capital of France?", "Paris is the capital of France."), ("What is the capital of France?",  "Berlin is the capital of Germany.") ] * 100
scores = np.array(list(model.rerank_pairs(pairs, parallel=2)))
scores
```
